```python
## Task 1: Activation functions: implement the softmax function. 5 points
import math


def sigm(z):
    return 1/(1 + np.exp(-z))

def dsigm(z):
    return sigm(z)*(1 - sigm(z))

def softmax(z):
    ''' softmax function for the output layer.
        The softmax function should be able to work if batch size is greater than 1.

        parameters:
            z: input numpy array. m_batch * 10
        
        return: m_batch * 10     
    '''
    ## add your code here
    
    #cache length of input vect
    batches = len(z)
    #init rslt vect
    y = np.empty( batches, dtype=list )
    
    #walk thru all input 
    for batchIndex in range(batches):
        #init z batch
        z_batch = z[batchIndex]
        #store size of batch
        batchSize = len(z_batch)
        #init rslt batch
        y_batch = np.empty( batchSize, dtype=float )
        #walk thru z_batch
        for i in range(batchSize):
            #calc desired output
            y_batch[i] = np.exp(z_batch[i]) / sum(np.exp(z_batch)) #sum(math.exp(z))
        #add batch to rslt vect
        y[batchIndex] = y_batch

    return y
    ##
    
# test the softmax function
z = np.array([[1, 2, 3], [4, 2, 4]])
s = softmax(z)
print(s)
```
